```python
# percentual de eleitores mulheres em relação ao total geral de eleitores
tem = round(((df['QTD_ELEITORES_FEMININO'].sum())/((df['QTD_ELEITORES'].sum()))*100),2)
```

```python
print(f'Percentual de mulheres eleitoras em relação ao total do eleitorado: {tem}%')
```

```python
# listar total de municípios onde o eleitorado masculino é maior que o eleitorado feminino
df[df['QTD_ELEITORES_MASCULINO'] > df['QTD_ELEITORES_FEMININO']].count()
```

```python
df['QTD_ELEITORES_FEMININO'] / df['QTD_ELEITORES_MASCULINO']
```

```python
df_BR = df[df['CD_PAIS']==1].copy()
```

```python
df_BR.head(6).T
```

```python
# relação com eleitorado feminino maior que o eleitorado masculino
df_BR['RELACAO_FM'] = df_BR['QTD_ELEITORES_FEMININO'] / df_BR['QTD_ELEITORES_MASCULINO']
```

```python
df_BR.nlargest(5, 'RELACAO_FM').T
```

```python
plt.figure(figsize=(15,5))
df_BR['RELACAO_FM'].plot.hist(bins=100, color='pink')
plt.title(f'eleitorado feminino x eleitorado masculino')
plt.grid();
```

```python
plt.figure(figsize=(15,5))
sns.displot(df_BR['RELACAO_FM'], bins=100, color='pink', kde=False)
plt.xlabel('eleitoras x eleitores')
plt.ylabel('Frequência')
plt.title(f'RELAÇÃO ELEITORADO F X M');
plt.axvline(1.0, color='black', linestyle='--')
```

```python
sns.swarmplot(data=df_BR, x='NM_REGIAO', y='RELACAO_FM')
plt.axhline(1.0, color='black', linestyle='--');
```
