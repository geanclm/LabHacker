# Data preparation

```python
# datagrame somente para análise do turno 2 e unidades dentro do Brasil
df_t2 = df[(df['NR_TURNO'] == 2) & (df['SG_UF'] != 'ZZ')]
```

```python
# quanidade de votos do candidato ao término do segundo turno
df_t2[df_t2['NM_CANDIDATO']=='JAIR MESSIAS BOLSONARO']['QT_VOTOS_NOMINAIS'].sum()
```

```python
df_t2[df_t2['NM_CANDIDATO']=='FERNANDO HADDAD']['QT_VOTOS_NOMINAIS'].sum()
```

```python
# dataframe com algumas colunas principais para simplificar a análise
df_v = df_t2[['SG_UF', 'CD_MUNICIPIO',
           'NM_MUNICIPIO','NR_CANDIDATO','NM_URNA_CANDIDATO','QT_VOTOS_NOMINAIS']]
```

```python
df_v.tail()
```

```python
df_v = df_v.groupby(['SG_UF','CD_MUNICIPIO','NM_MUNICIPIO','NR_CANDIDATO','NM_URNA_CANDIDATO']).sum().reset_index()
```

```python
df_v.sample(5)
```

```python
df_v[df_v['NM_MUNICIPIO'] == 'BELO HORIZONTE'].tail()
```

```python
# dataframe ordenado por quantidade de votos de cada munícipio
df_v.sort_values(by='QT_VOTOS_NOMINAIS', ascending=False).head(15)
```

```python
# salvar listagem diretamente no datagrame df_v
df_v.sort_values(by='QT_VOTOS_NOMINAIS', ascending=False, inplace=True)
```

```python
# excluir linhas duplicadas do dataframe df_v para deixar somente o resultado
# do candidato mais votodade em cada município
df_v.drop_duplicates(subset='CD_MUNICIPIO', keep='first', inplace=True)
```

```python
df_v.head(10)
```

```python
df_v[df_v['NM_MUNICIPIO'] == 'BELO HORIZONTE'].tail()
```

```python
# criar coluna boleana para representar o NR_CANDIDATO de acordo com resultado do candidato
# Se o candidato venceu no município o registro do campo deve ser 'true'
df_v['RESULTADO'] = df_v['NR_CANDIDATO'].apply(lambda x : x==17)
```

```python
# Desafio encontrado:
#     Integrar os registros do .csv com as votações e os dados dos shapefile de geololização
```

# Data preparation for the map

```python
# fazer o merge pelo index dos dataframes
# definir o índice de cada dataframe com a coluna autêntica (código único)
```

```python
df_v2 = df_v.copy()
```

```python
df_v2.set_index('CD_MUNICIPIO',drop=False, inplace=True)
```

```python
df_v2.loc[(df_v2['NM_MUNICIPIO']=='OURO BRANCO')]
```

```python
df_v2.info()
```

```python
df_estadao2 = df_estadao.copy()
```

```python
df_estadao2.set_index('COD_TSE',drop=False, inplace=True)
```

```python
df_estadao2.loc[(df_estadao2['NOME']=='OURO BRANCO')]
```

```python
df_estadao2.info()
```

```python
# Merge do datagrame df_v2 com o df_estadado2
# Obs: olhar na documentação mais informações sobre o join
df_v2_estado2 = df_v2.join(df_estadao2)
```

```python
df_v2_estado2.loc[(df_v2_estado2['NOME']=='OURO BRANCO')]
```

```python
# Merge do dataframe df_v2_estado2 com o df_map.loc (geopandas)
```

```python
df_v2_estado2.info()
```

```python
df_map.info()
```

```python
# a coluna 'GEOCOD_IBGE' de código do datagrame df_v2_estado2
# precisa ser do mesmo tipo de dados da coluna equivalente
# no dataframe df_map 'CD_MUN'
df_v2_estado2['GEOCOD_IBGE'] = df_v2_estado2['GEOCOD_IBGE'].astype(str)
```

```python
df_v2_estado2.info()
```

```python
# aterar o índice de cada dataframe para ter uma coluna em comum

# 1 - mudar o índice do dataframe df_map
df_map.set_index('CD_MUN', drop=False, inplace=True)

# 2 - mudar o índice do datagrame
df_v2_estado2.set_index('GEOCOD_IBGE', drop=False, inplace=True)
```

```python
# df_map.head()
df_map.loc[(df_map['NM_MUN']=='Ouro Branco')]
```

```python
# df_v2_estado2.head()
df_v2_estado2.loc[(df_v2_estado2['NM_MUNICIPIO']=='OURO BRANCO')]
```

```python
# Merge dos datafranes df_map e df_v2_estado2
df_map_v2_estado2 = df_map.join(df_v2_estado2)
```

```python
df_map_v2_estado2.head().T
```

```python
df_map_v2_estado2.info()
```

```python
# localizar valores nulos
# df_map_v2_estado2[df_map_v2_estado2['AJUSTE'].isnull()].T
```

```python
# remover a coluna 'AJUSTE' e dados ausentes
df_map_v2_estado2.drop(['AJUSTE'],axis = 1, inplace=True)
```

```python
# Referência oara o código de cor do mapa
# False/0 = red, True/2, green
from matplotlib.colors import ListedColormap
```

```python
cmap = ListedColormap(['red','green'])
```

```python
# Mapa
fig, ax = plt.subplots(1, figsize=(15,15))

ax = df_map_v2_estado2.plot(column='RESULTADO',
                            cmap=cmap, legend=True, linewidth=0.1, ax=ax, edgecolor='grey')

# ocultar os eixos do mapa
ax.set_axis_off()

# Título do mapa
ax.set_title('Resultado Eleição 2018 / Município / 2º Turno',
             fontdict={'fontsize':'25', 'fontweight':'3'})

# Rodapé
ax.annotate('Fonte IBGE 2015; TSE, 2019; Jornal o Estado de São Paulo, 2019',
    xy=(0.1, 0.08), xycoords='figure fraction',
    horizontalalignment='left',
    verticalalignment='top', fontsize=10, color='#555555')
    
# Alterar a legenda
leg = ax.get_legend()
# leg.get_texts()[0].set_text('Fernando Haddad')
# leg.get_texts()[1].set_text('Jair Bolsonaro')

plt.show()
```
