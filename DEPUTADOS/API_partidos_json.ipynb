# leitura de dados abertos da Câmara
## Análise de dados obtidos pela API da Câmara dos Deputados
### fonte de dados: https://dadosabertos.camara.leg.br/swagger/api.html

```python
import pandas as pd
import requests as rq
```

```python
url = 'https://dadosabertos.camara.leg.br/api/v2/partidos?itens=100&ordem=ASC&ordenarPor=sigla'
```

```python
response = rq.get(url)
```

```python
# response.text
```

```python
df = pd.DataFrame(response.json()['dados'])
```
