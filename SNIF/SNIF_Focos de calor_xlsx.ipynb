# Import libs

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Import data

```python
# fonte: https://snif.florestal.gov.br/pt-br/incendios-florestais
```

```python
df = pd.read_excel('focos_calor_1998_2019.xlsx')
```

```python
df.shape
```

```python
df.info()
```

```python
df
```

```python
df[df['Número']==df['Número'].max()]
```

```python
df2 = df[(df['Mês']!='Total')&
       (df['Ano']!='Máximo')&
       (df['Ano']!='Média')&
       (df['Ano']!='Mínimo')&
       (df['Período']!='Máximo')&
       (df['Período']!='Média')&
       (df['Período']!='Mínimo')]
```

```python
df2[df2['Número']==df2['Número'].max()]
```

# Data Graph

```python
df_2007 = df2[df2['Ano']==2007]
```

```python
df_2007.plot(x='Mês',y='Número')
plt.title(f'Focos de Calor Ano 2007')
plt.grid()
```

```python
df_2007.plot.bar(x='Mês',y='Número')
plt.title(f'Focos de Calor Ano 2007')
plt.grid()
```

```python
plt.figure(figsize=(15,5))
plt.plot('Mês','Número', data=df2[df2['Ano']==2016], label='2016')
plt.plot('Mês','Número', data=df2[df2['Ano']==2017], label='2017')
plt.plot('Mês','Número', data=df2[df2['Ano']==2018], label='2018')
plt.legend();
plt.title('Focos de Calor')
plt.xlabel('Mês')
plt.ylabel('Quantidade')
plt.xticks(rotation=45)
plt.legend(fontsize=12)
plt.grid();
```

```python
# Gráfico om o Seaborn
plt.figure(figsize=(15,5))
sns.barplot(x='Mês',y='Número', data=df_2007)
plt.title('Focos de Calor')
plt.xticks(rotation=45)
plt.grid();
```
